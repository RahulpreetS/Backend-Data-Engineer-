```python
import numpy as np
import pandas as pd
import re
import os
import glob
```

```python
# MAKE A LIST OF ALL EXCEL FILES IN THE DATA FOLDER
import os, fnmatch
pattern="*.xlsx" # pattern to match excel files
inputPath="/Users/savina/MOOCS BullsEye/Data Science/Amit/Data/"

dataframes_Detail_67 = []
dataframes_DetailVol_67 = []
dataframes_DetailTemp_67 = []

for base, dirs, files in os.walk(inputPath):
        for name in files:
            if(fnmatch.fnmatch(name,pattern)):
                inputFile = os.path.join(base,name)
                print(inputFile)
                file = pd.ExcelFile(inputFile)
                for sheet in file.sheet_names:
                    if re.match('Detail_67_*', sheet):  # when matching pattern add the dataframe to the list
                        print(sheet, ",", inputFile)
                        dataframes_Detail_67.append(pd.read_excel(inputFile, sheet_name=sheet)) # MAKE A LIST OF ALL SHEETS WITH NAME STARTING FROM Detail_67_
                        #print(dataframes_Detail_67)
                    elif re.match('DetailVol_67_*', sheet):  # when matching pattern add the dataframe to the list
                        print(sheet, ",", inputFile)
                        dataframes_DetailVol_67.append(pd.read_excel(inputFile, sheet_name=sheet)) # MAKE A LIST OF ALL SHEETS WITH NAME STARTING FROM DetailVol_67_
                    elif re.match('DetailTemp_67_*', sheet):  # when matching pattern add the dataframe to the list
                        print(sheet, ",", inputFile)
                        dataframes_DetailTemp_67.append(pd.read_excel(inputFile, sheet_name=sheet))# MAKE A LIST OF ALL SHEETS WITH NAME STARTING FROM DetailTemp_67_
#print(dataframes_DetailTemp_67)
Detail_67_df = pd.concat(dataframes_Detail_67).drop_duplicates('Record Index').reset_index(drop=True)
Detail_67_df.to_csv('detail.csv')

DetailVol_67 = pd.concat(dataframes_DetailVol_67).drop_duplicates('Record ID').reset_index(drop=True)
DetailVol_67.to_csv('detailVol.csv')

DetailTemp_67 = pd.concat(dataframes_DetailTemp_67).drop_duplicates('Record ID').reset_index(drop=True)
DetailTemp_67.to_csv('detailTemp.csv')

    
```

```python
# Downsampling to 1 min.

import os, fnmatch
pattern="*.csv" # pattern to match excel files
inputPath="/Users/savina/MOOCS BullsEye/Data Science/Amit/Data/"

for base, dirs, files in os.walk(inputPath):
        for name in files:
            if(fnmatch.fnmatch(name,pattern)):
                inputFile = os.path.join(base,name)                                
                if(name == 'detail.csv'):
                    print("HEllo")
                    df = pd.read_csv(inputFile, parse_dates=['Absolute Time'])
                    print(df)
                    sampleddf = df.resample('T',on='Absolute Time').mean()
                    print(sampleddf)
                    sampleddf.to_csv('detailDownsampled.csv')
                elif(name == 'detailVol.csv'):
                    df = pd.read_csv(inputFile, parse_dates=['Realtime'])
                    sampleddf = df.resample('T',on='Realtime').mean()
                    sampleddf.to_csv('detailVolDownsampled.csv')
                elif(name == 'detailTemp.csv'):
                    df = pd.read_csv(inputFile, parse_dates=['Realtime'])
                    sampleddf = df.resample('T',on='Realtime').mean()
                    sampleddf.to_csv('detailTempDownsampled.csv')
                    
```
